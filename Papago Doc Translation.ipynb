```python
import os
from dotenv import load_dotenv

# .env 파일 경로 설정 (Colab 기본 경로는 /content)
env_path = '/content/.env'

# .env 파일 로드
load_dotenv(env_path)

# 환경 변수 사용
api_key = os.getenv('API_KEY')
secret_key = os.getenv('SECRET_KEY')

print(f"API_KEY: {api_key}")
print(f"SECRET_KEY: {secret_key}")
```

## Papago Doc Translation

### 문서 번역

```python
import requests
from requests_toolbelt import MultipartEncoder
import uuid

file_path = '/content/dong1.docx'

data = {
  'source': 'zh-CN',
  'target': 'ko',
  'file': ('dong1.docx', open(file_path, 'rb'), 'application/octet-stream', {'Content-Transfer-Encoding': 'binary'})
}
m = MultipartEncoder(data, boundary=uuid.uuid4())

headers = {
  "Content-Type": m.content_type,
  "X-NCP-APIGW-API-KEY-ID": api_key,
  "X-NCP-APIGW-API-KEY": secret_key
}

url = "https://naveropenapi.apigw.ntruss.com/doc-trans/v1/translate"
res = requests.post(url, headers=headers, data=m.to_string())
print(res.text)
```

### 문서 번역 상태 확인

```python
import requests

RequestId = "20250105_1736081930267_078"
headers = {
  "X-NCP-APIGW-API-KEY-ID": api_key,
  "X-NCP-APIGW-API-KEY": secret_key
}

url = "https://naveropenapi.apigw.ntruss.com/doc-trans/v1/status?requestId=" + RequestId
res = requests.get(url, headers=headers)
print(res.text)
```

### 문서 번역 결과 다운로드

```python
import urllib.request

url = "https://naveropenapi.apigw.ntruss.com/doc-trans/v1/download?requestId=" + RequestId

opener = urllib.request.build_opener()
opener.addheaders = [('X-NCP-APIGW-API-KEY-ID', api_key), ('X-NCP-APIGW-API-KEY', secret_key)]
urllib.request.install_opener(opener)

urllib.request.urlretrieve(url, "dongt1.docx")
```
