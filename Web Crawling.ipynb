## 2 . 웹 페이지 가져오기


```python
import requests

# 크롤링할 웹페이지 URL
url = "https://app.mediclassics.kr/books/동의보감/pages/2167"  # 실제 URL로 변경

# 페이지 요청
response = requests.get(url)

# HTML 내용 확인
html_content = response.text
```

## 3 . BeautifulSoup으로 HTML 파싱

```python
from bs4 import BeautifulSoup

# HTML 파싱
soup = BeautifulSoup(html_content, 'html.parser')

# class="KO"에 해당하는 모든 텍스트 추출
ko_elements = soup.find_all(class_="KO")

# 텍스트만 출력
for element in ko_elements:
    print(element.get_text(strip=True))
```

## 4 . 결과 저장


```python
# 텍스트를 .txt 파일로 저장
with open('ko_texts.txt', 'w', encoding='utf-8') as file:
    for element in ko_elements:
        file.write(element.get_text(strip=True) + '\n')

print("텍스트가 ko_texts.txt 파일에 저장되었습니다.")
```

## 5 . 동의보감.txt

```python
import requests
from bs4 import BeautifulSoup
import time  # 텀을 주기 위해 time 모듈 추가

# 결과 저장할 파일 열기
with open('동의보감.txt', 'w', encoding='utf-8') as file:
    # 1부터 2888까지 반복
    for page in range(1, 2889):
        # 크롤링할 URL
        url = f"https://app.mediclassics.kr/books/동의보감/pages/{page}"

        # 웹페이지 요청
        response = requests.get(url)

        # 요청 성공 시 처리
        if response.status_code == 200:
            # HTML 파싱
            soup = BeautifulSoup(response.text, 'html.parser')

            # class="KO"에 해당하는 모든 텍스트 추출
            ko_elements = soup.find_all(class_="KO")

            # 텍스트 파일에 저장
            for element in ko_elements:
                text = element.get_text(strip=True)
                file.write(text + '\n')
        else:
            # 요청 실패 시 로그 출력
            print(f"페이지 {page}에서 요청 실패 (상태 코드: {response.status_code})")

        # 요청 간 1초 간격 추가
        time.sleep(1)

print("모든 페이지 크롤링이 완료되었습니다. 텍스트가 동의보감.txt 파일에 저장되었습니다.")
```
